## 3. Filling the Missing Values – Imputation

In this case, we will be filling the missing values with a certain number.

The possible ways to do this are:
- Filling the missing data with the **mean** or **median** values ( if it's a nautural variavle)
- Filling the missing data with **mode** if it’s a **categorical value**.
- Filling the numerical value with **0 or -999**, or **some other number** that will not occur in the data.This can be done so that the machine can recognize that the data is not real or is different.
- Filling the categorical value with a **new type** for the missing values.

```python
# df['Company Rating'].mean() is used to get the mean of all the values in the a particular column
df['Company Rating'].mean()
```

```python
df['Company Rating'].median()
# df['Company Rating'].median() is used to get the median of all the values in the a particular column
```

- __fillna: fills the null records__
- __dropna: drops the null records__

#### 1. Filling the missing values with the mean value


```python
#fillna: fills the null records
updated_df = df
updated_df['Company Rating']=updated_df['Company Rating'].fillna(df['Company Rating'].mean())
updated_df.info()
```

**fillna()**

#### 4. Forward & Backward Filling – Imputation

**Forward and backward** filling are common techniques used in exploratory data analysis (EDA) to handle missing data. These methods are often used when dealing with time series data or any dataset where the order of observations is important. Here’s a brief overview of both techniques:

**Forward Filling (ffill)**
Forward filling propagates the last observed non-missing value forward to fill missing values. This method is useful when you believe that the most recent non-missing value is a good estimate for the missing values that follow.

**Backward Filling (bfill)**
Backward filling propagates the next observed non-missing value backward to fill missing values. This method is useful when you believe that the future non-missing value is a good estimate for the missing values that precede it.

```python
df_updated2=pd.read_csv("Data-Science-Job_Listing.csv")
```

```python
df_updated2.info()
```

```python
df_updated2['Company Rating']=df_updated2['Company Rating'].ffill()
```

```python
df_updated2.info()
```

```python
#using mode()
df2=df
```

```python
df2['Salary']=df2['Salary'].fillna(df2['Salary'].mode()[0])
df2.info()
```

The '[0]' is used to extract the first element of the mode result. The mode() method in Pandas returns a Series containing all the mode(s) of the column. If there are multiple modes (i.e., multiple values that occur with the same highest frequency), the mode() method will return all of them. By using [0], you are selecting the first mode from this Series.

Here's a step-by-step breakdown:

df2['Salary'].mode() returns a Series of the most frequent value(s) in the 'Salary' column.
[0] selects the first element of this Series, which is the first mode.
