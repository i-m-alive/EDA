```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df=pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=["CustomerId","Surname","RowNumber"],axis=1, inplace=True)
```

```python
df.head()
```

#### Lable Encoding

Its is best suits for ordinal values but we are doing here for nominal data

```python
le=preprocessing.LabelEncoder()
df['Gender_Label']=le.fit_transform(df.Gender.values)
```

```python
df.head()
```

```python
df.sample(5)
```

### One Hot Encoding

We can do this in two ways
1. from sklearn.preprocessing import OneHotEncoder
2. get_dummies

```python
one_hot= pd.get_dummies(df["Geography"])
one_hot.head()
```

```python
#to convert the complete data set into one hot encoding
df.drop(columns=['Gender_Label'],axis=1,inplace=True)
df_dummies=pd.get_dummies(df)
df_dummies.head()
```

To avoid the dummy variable trap, a common practice is to drop one of the one-hot encoded columns, which reduces the number of variables by one and removes the linear dependency. This is equivalent to choosing a reference category and comparing the other categories to it.

```python
df_dummies_new=pd.get_dummies(df,drop_first=True)
df_dummies_new.head()
```
