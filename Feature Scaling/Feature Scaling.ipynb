Two types of Feature Sacling are:
1. Normalization
2. Standardization

**Normalization**

- It is also called **MIN-MAX** Feature Sacling.
- In this technique we use a formula to rescale the value between 0-1.
- i.e.. X'=(X-X(min))/X(max)-X(min)

**Standardization**

- It is also called **Z-score**
- Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation.
- formula for this is : x' = (x-mean)/S.D

### NORMALIZATION

```python
new_df=pd.DataFrame(df,columns=['Age','Absences'])
```

```python
new_df.head()
```

```python
new_df.info()
```

```python
new_df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler is used for normalization
```

```python
scaler=MinMaxScaler() 
normalized_df = scaler.fit_transform(new_df)
```

```python
print(normalized_df)
```

it uses the formula of min-max(ie.. X=(X-Xmin)/(Xmax-Xmin)

### Standardization or Z-Score

```python
from sklearn.preprocessing import StandardScaler
```

```python
s_scaler=StandardScaler()
```

```python
standard=s_scaler.fit_transform(new_df)
print(standard)
```

if uses the form of standared scaler(i.e..  X=(X-Mean)/S.D)
